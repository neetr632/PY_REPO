```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

iris = load_iris()
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df['species'] = iris.target

df['species'] = df['species'].map({0:'setosa', 1:'versicolor', 2: 'virginica'})

# Plot Sepal Length vs Sepal Width
plt.figure(figsize=(10, 6))
for species in df['species'].unique():
    subset = df[df['species'] == species]
    plt.plot(subset['sepal length (cm)'],subset['sepal width (cm)'],marker='o', linestyle='', label=species)

plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.title('Sepal Length vs Sepal Width for Iris Species')
plt.legend(title='Species')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
dataset  = pd.read_csv('/content/headbrain.csv')
dataset.head()
```

```python
dataset.isnull().sum()
```

```python
dataset.duplicated().sum()
```

```python
y = dataset[['Gender']].values
print(y)
```

```python
x = dataset[['Age Range','Head Size(cm^3)','Brain Weight(grams)']].values
print(x)
```

```python
import seaborn as sea
import matplotlib.pyplot as plt
sea.heatmap(dataset.corr(),annot=True)
plt.show()
```

```python
import numpy as np
x_mean = np.mean(x)
y_mean = np.mean(y)
# print(x_mean, y_mean)

n = len(x)
numer = 0
denome = 0
for i in range(n):
    numer += (x[i] - x_mean) * (y[i] - y_mean)
    denome += (x[i] - x_mean) ** 2

b1 = numer / denome
b0 = y_mean - (b1 * x_mean)
print("Coeffient Values: ",b1,b0)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize']= (10.0,5.0)

y_pred = b0 + b1 * x
plt.plot(x, y_pred, color="Red", label="Regression Line")
plt.scatters(x, y, color="yellow", label="Scatter Data")
plt.xlabel('Head Size in cm3')
plt.ylabel('Brain Weight in grams')
plt.legend()
plt.show()
```

```python
#calculating the root mean square error
rmse = 0
for i in range(n):
    y_pred = b0 + b1 * x[i]
    rmse += (y_pred - y[i]) ** 2

rmse = np.sqrt(rmse/n)
print("Root mean Sqaure error is ", rmse)
```
