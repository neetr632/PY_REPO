```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler, LabelEncoder

# Convert data to DataFrame
x = pd.DataFrame(dataset)

# Drop the 'Loan_ID' and 'Target' columns
x = x.drop(columns=['Loan_ID', 'Target'])

# Define numeric and categorical columns
numeric_cols = ['ApplicantIncome', 'CoapplicantIncome', 'LoanAmount', 'Loan_Amount_Term', 'Credit_History']
categorical_cols = ['Gender', 'Married', 'Dependents', 'Education', 'Self_Employed', 'Property_Area']

# Handle missing values for numeric columns
x[numeric_cols] = x[numeric_cols].fillna(x[numeric_cols].mean())

# Encode categorical columns using LabelEncoder
le = LabelEncoder()
for col in categorical_cols:
    x[col] = le.fit_transform(x[col].astype(str))  # Convert to string if not already

# Apply StandardScaler to numeric columns
sc = StandardScaler()
x[numeric_cols] = sc.fit_transform(x[numeric_cols])

# Display the processed dataset
print("Processed and scaled x dataset:")
print(x.head())
```
