# **1. What Are Tokens and Spans in NLP?**

1. tokens are individual words or punctuation marks, but the definition of a token can vary based on the language model.

2.  in spaCy, tokenization involves splitting the text into words, punctuation, numbers, or symbols. SpaCy uses pre-trained models to define what constitutes a token, based on language rules and patterns (e.g., handling contractions, splitting sentences on spaces, punctuation, etc.).

**2. How Does SpaCy Determine Tokens?**
* Language-Specific Rules:
 spaCy includes language-specific rules for splitting text into tokens. For example:

* In English, spaces are often used to split words.
Punctuation marks like periods, commas, and quotes are treated as separate tokens.
* Special cases like abbreviations (e.g., "Dr.") and contractions (e.g., "isn't") are handled by specific rules in the tokenizer.

# **4. Try tokenizing "NEW YORK" – you'll end up with two separate tokens instead of one, which is not what you want. If tokenization separates a name into two tokens, doesn't it seem illogical since entity names should be written together? How does spaCy fix this abnormality?**

* In spaCy, proper nouns like "New York" are typically split into individual tokens by default, even though they represent a single entity (a named entity). This is because the tokenizer treats each word separately unless explicitly instructed otherwise. So, "New York" is split into two tokens: "New" and "York".

* However, while "New York" is tokenized into two separate tokens, spaCy also performs ***named entity recognition (NER)***, which recognizes "New York" as a single entity (a location in this case).

# **5. Named Entity Recognition (NER)**

* **How Named Entity Recognition Works**

1. **Detection**: The model first detects a sequence of words (called "entities") in the text that represents something of interest.

2. **Classification**: After detecting the entities, it classifies them into categories like:
* **Person (PER)**: Names of people (e.g., "John Doe").
* **Organization (ORG)**: Names of companies, institutions, or organizations (e.g., "Google").
* **Location (LOC/GPE)**: Geographical locations like cities, countries, or regions (e.g., "New York").
* **Date/Time (DATE/TIME)**: Dates, times, or durations (e.g., "January 1, 2023").
3. **Miscellaneous:** Other Types of entities like products, events, or artworks.

```python
def count_up_to(max):
    count = 1
    while count <= max:
        yield count
        count += 1
result = count_up_to(5)
print(type(result))
for num in result:
  print(num)
```

* Similar to right_edge, the left_edge represents the token that lies furthest to the left within the syntactic subtree that is headed by token_1.

**In simpler terms:**

# The syntactic subtree consists of token_1 and all the tokens connected to it (its syntactic dependents).

* The left_edge is the token that is farthest to the left in this structure.
If token_1 is not part of a larger structure (i.e., it has no dependents), the left_edge will be token_1 itself.

# **token_1.ent_iob_**
is an attribute in spaCy that indicates whether a token is part of an entity and, if so, its position within that entity. It stands for Inside-Outside-Beginning (IOB) tagging and provides information about the token's relationship to named entities in the text.

The possible values are:

* O: The token is outside of a named entity.
* B: The token is at the beginning of a named entity.
* I: The token is inside a named entity but not at the beginning.

### Doubt:

In spaCy, if a phrase like "from the disruption or misdirection of the services they provide" is a multi-token span, why might an individual token like "the" have the same **left_edge** and **right_edge?**

### Answer:

When you pick a token within a larger span, its `**left_edge**` and `**right_edge**` will be the same because it is considered an individual token. However, if you select the starting token of the larger span, the `**left_edge**` and `**right_edge**` will show different values, reflecting the boundaries of the entire span. This is because the starting token marks the beginning of the span, whereas the token inside the span only pertains to its own boundaries.
